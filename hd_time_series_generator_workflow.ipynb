```python
import pandas as pd
import numpy as np
import math
import os
import hd_time_series_generator 
import datetime 
import importlib
import geopandas as gpd
from data import *
import time
import matplotlib.pyplot as plt
```

```python
# load the config file
config = load_config()

#define the case study name
case_study = "Frauental"
```

```python
# load the transition matrix
path = os.path.join(case_study,config["parameter_dir"],config["MCMC_dir"])
df_transition_matrix_WD = pd.read_excel(os.path.join(path, config["case_study_data"]["transition_matrix_WD"]), index_col=0)
df_transition_matrix_WE = pd.read_excel(os.path.join(path, config["case_study_data"]["transition_matrix_WE"]), index_col=0)

# load the temperature data
df_temperature = hd_time_series_generator.prepare_temp_out(case_study, config)

df_temperature.head()
```

```python
# generate the time sereies for all buildings
gdf_buildings = gpd.read_file(os.path.join(case_study, config["building_data"]["buildings_gdf"]))

# Call the function
df_buidling_TS = hd_time_series_generator.generate_building_time_series(gdf_buildings, df_temperature, df_transition_matrix_WD, df_transition_matrix_WE)
df_buidling_TS.head()
```

```python
# save time series to disk
hd_time_series_generator.write_hdts_to_disk(df_buidling_TS, case_study, config)
```

```python
# get the total heating demand over all buildings and hours in GWh
total_heating_demand = sum(df_buidling_TS.iloc[:, 1:].sum()) / 1e6
print(f"Total heating demand: {total_heating_demand:.2f} GWh")
```
